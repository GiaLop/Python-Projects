# Analisi Piloti F1
Implementazione delle funzioni necessarie per ottenre le seguenti statistiche:
- in relazione all'input una lista di: tot punti, gp vinti, gp a podio
- un dizionario con la classifica pilota:punteggio
- un dizionario con la classifica costruttore:punteggio

```python
# Importo la libreria csv per leggere il file su cui lavorare
import csv

file = "/content/drive/MyDrive/Profession AI/formula1_data.csv"

def open_file(file):

  """

  Funzione per aprire agevolmete il file nelle funzioni seguenti

  """

  with open(file) as f:
    return list(csv.DictReader(f))

def count_driver_positions(driver):

  """

  Contare le posizioni ottenute dal pilota

  """

  f1 = open_file(file)
  total = []
  for row in f1: # itero sulle righe del file
    if row["Driver"] == driver and row["Position"] != 0: # match tra pilota e posizione
      total.append(int(row["Position"])) # aggiungo la posizione alla lista
  return total

def points_position(position):

  """

  Calcolare in relazione alle posizioni il punteggio finale

  """

  if position == 1:
    return 10
  elif position == 2:
    return 8
  elif position == 3:
    return 6
  elif position == 4:
    return 5
  elif position == 5:
    return 4
  elif position == 6:
    return 3
  elif position == 7:
    return 2
  elif position == 8:
    return 1
  else:
    return 0

def total_points(total):

  """

  Somma punteggio in relazione posizioni e punti per piazzamento

  """

  tot_points = 0
  for p in total: # itero sulla lista
    points = points_position(p) # calcolo il punteggio per posizione
    tot_points+=points # aggiorno il punteggio totale
  return tot_points

def count_wins(driver):

  """

  Contiamo solo quando il pilota si è piazzato primo

  """

  f1 = open_file(file)
  tot_wins = 0
  for row in f1: # itero su righe file
    if row["Driver"] == driver and int(row["Position"]) == 1: # se il pilota si è piazzato primo
      tot_wins+=1 # aggiorno il counter con solo le prime posizioni
  return tot_wins

def top_3(driver):

  """

  Contiamo tutte le volte che il pilota è salito sul podio

  """

  f1 = open_file(file)
  top_3_count = 0
  for row in f1:
    if row["Driver"] == driver and int(row["Position"]) in [1, 2, 3]:
      top_3_count+=1 #aggiorno il counter solo se il pilota è sul podio
  return top_3_count

# Inglobo le funzioni in un'unica def la quale ritorna la prima statistica

def pilot_info(driver):
  all_info = []
  positions = count_driver_positions(driver)
  total_pt = all_info.append(total_points(positions))
  wins = all_info.append(count_wins(driver))
  podiums = all_info.append(top_3(driver))
  return all_info

# Definisco la funzione per la seconda statistica
# Apro un file nuovo nel quale scrivo i risultati della classifica

def calculate_driver_rank(file):

  f1 = open_file(file) # apro il file
  final_rank = {} # inizializzo un dict vuoto
  for row in f1:
    nome_pilota = row["Driver"] # definisco chiave
    positions = count_driver_positions(nome_pilota) # definisco valore
    points = total_points(positions) # calcolo
    final_rank[nome_pilota] = points # e aggiorno il dizionario con il risultato
  return final_rank

with open("Drivers_Standings_2008.txt", "w") as f1_2008: # creo il file nuovo
  f1_2008.write("Drivers Standings 2008 Formula 1") # scrivo l'header
  f1 = open_file(file) # apro il file da cui prendo le infos
  f_rank = calculate_driver_rank(file) # calcolo il rank
  for key, values in f_rank.items(): # scrivo le key:value nel file creato
    f1_2008.write("\n"+ key +": "+ str(values))

# Definisco la funzione per la terza statistica

def count_team_positions(team):

  f1 = open_file(file)
  total = []
  for row in f1: # itero sulle righe del file
    if row["Team"] == team and row["Position"] != 0: # match tra pilota e posizione
      total.append(int(row["Position"]))
  return total

def rank_team(file): # stesso procedimento dei drivers ma con teams

  f1 = open_file(file)
  constructors_rank = {}
  for row in f1:
    nome_costruttore = row["Team"]
    position = count_team_positions(nome_costruttore)
    points = total_points(position)
    constructors_rank[nome_costruttore] = points
  return constructors_rank

# SVOLGIMENTO

# Controlliamo se il pilota è nella nostra lista
while True:
  sel_driver = input("Scegli un pilota: ")
  f1_data = open_file(file)
  nomi = [row["Driver"] for row in f1_data]
  if sel_driver in nomi:
    break
  else:
    print("Pilota non trovato")

# Invoco le funzioni per ottenere le statistiche richieste
info = pilot_info(sel_driver)
drivers_rank = calculate_driver_rank(file)
rank_t = rank_team(file)
print(info)
print(drivers_rank)
print(rank_t)
```
